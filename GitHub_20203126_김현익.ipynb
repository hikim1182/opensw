```python
import cv2
from google.colab.patches import cv2_imshow
path="/content/drive/MyDrive/Colab Notebooks/cv2 사진/치킨.png"
image=cv2.imread(path, cv2.IMREAD_COLOR)
cv2_imshow(image)
```

```python
# 이미지를 그레이스케일로 변환
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 노이즈를 줄이고 에지 검출을 돕기 위해 가우시안 블러 적용
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Canny 에지 검출 적용
edges = cv2.Canny(blurred, 50, 150)

# 에지 표시
cv2_imshow(edges)
```

1. Canny 에지 검출 알고리즘이 그레이스케일 이미지에서 작동하기 때문에 이미지를 그레이스케일로 변환

2. 가우시안 블러가 그레이스케일 이미지에 적용되어 노이즈를 감소시키고 에지 검출의 성능을 향상

3. Canny 에지 검출이 블러 처리된 이미지에 적용

4. 에지 표시
